### Aim: To detect edges using Canny Edge Detection algorithm

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

def ImageDisplayFunc(Image):
    plt.figure(figsize = (16, 12))
    plt.imshow(Image, cmap = 'gray')
    plt.show()
    
img = cv.imread('C:\\Users\\ASHOK\\Desktop\\Python files\\Image Processing\\Images\\Berries.png')

GrayscaleImg = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

ImageDisplayFunc(GrayscaleImg)
```

## Using Canny Edge Detection algorithm

```python
ProcessedImage = cv.Canny(GrayscaleImg, 80, 180)

ImageDisplayFunc(ProcessedImage)
```
