### Aim: To perform image enhancement using Spatial Domain filters

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

def ImageDisplayFunc(Image):
    plt.figure(figsize = (16, 12))
    plt.imshow(Image)
    plt.show()
```

## Gaussian filter

### Image to be enhanced:

```python
img = cv.imread('C:\\Users\\ASHOK\\Desktop\\Python files\\Image Processing\\sample noisy images\\image3.jpg')

ImageDisplayFunc(img)
```

```python
ProcessedImage = cv.GaussianBlur(img, (5, 5), 0)
```

### Enhanced or restored image:

```python
ImageDisplayFunc(ProcessedImage)
```

## Median filter (removes salt and pepper noise well)

```python
ProcessedImage = cv.medianBlur(img, 5)
```

### Enhanced image:

```python
ImageDisplayFunc(ProcessedImage)
```

## Mean filter

```python
ProcessedImage = cv.blur(img, (5, 5))
```

### Enhanced image:

```python
ImageDisplayFunc(ProcessedImage)
```
