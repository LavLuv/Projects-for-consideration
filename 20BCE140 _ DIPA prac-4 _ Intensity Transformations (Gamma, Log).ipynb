### Aim: To implement Gamma, Log and Inverse Log transformations (for Grayscale images)

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

def ImageDisplayFunc(Image, ImageSizeTuple):
    plt.figure(figsize = ImageSizeTuple)
    plt.imshow(Image)
    plt.show()
```

### Original Grayscale image

```python
img = cv.imread('C:\\Users\\ASHOK\\Desktop\\Python files\\Image Processing\\Grayscale image 2.png')
```

```python
ImageDisplayFunc(img, (16, 12))
```

## Gamma transformation

```python

# r: pixel value of original Grayscale image
# s: pixel value of processed Grayscale image

"""
    1. Gamma transformation
    
        s = c * (r ^ gamma)
"""

gamma = float(input('-> Enter the value of gamma: '))
print('')

c = float(input('-> Enter the value of c: '))
print('')

def GammaTransform(image, gamma, c):
    ImageMatrix = list(image)
    ProcessedImageMatrix = []
    for ImageRow in ImageMatrix:
        ProcessedImageRow = []
        for pixel in ImageRow:
            ProcessedPixel = []
            r = pixel[0]
            s = round(c * np.power(r, gamma))
            for i in range(0, 3):
                ProcessedPixel.append(s)
            ProcessedImageRow.append(ProcessedPixel)
        ProcessedImageMatrix.append(ProcessedImageRow)
    ProcessedImage = np.array(ProcessedImageMatrix)
    
    return ProcessedImage


ProcessedImage = GammaTransform(img, gamma, c)
```

### Transformed Grayscale image

```python
ImageDisplayFunc(ProcessedImage, (16, 12))
```

## Log transformation

```python

# r: pixel value of original Grayscale image
# s: pixel value of processed Grayscale image

"""
    1. Log transformation
        
            s = c * ln(r + 1)
""" 

c = float(input('-> Enter the value of c: '))
print('')

def LogTransform(image, c):
    ImageMatrix = list(image)
    ProcessedImageMatrix = []
    for ImageRow in ImageMatrix:
        ProcessedImageRow = []
        for pixel in ImageRow:
            ProcessedPixel = []
            r = pixel[0]
            s = round(c * np.log(r + 1))
            for i in range(0, 3):
                ProcessedPixel.append(s)
            ProcessedImageRow.append(ProcessedPixel)
        ProcessedImageMatrix.append(ProcessedImageRow)
    ProcessedImage = np.array(ProcessedImageMatrix)
    
    return ProcessedImage
```

```python
ProcessedImage = LogTransform(img, c)
```

### Transformed Grayscale image

```python
ImageDisplayFunc(ProcessedImage, (16, 12))
```

## Inverse Log transformation

```python

# r: pixel value of original Grayscale image
# s: pixel value of processed Grayscale image

"""     
        1. Inverse Log transformation
        
                s = c * (1 / ln(r + 1))
"""

c = float(input('-> Enter the value of c: '))
print('')

def InverseLogTransform(image, c):
    ImageMatrix = list(image)
    ProcessedImageMatrix = []
    for ImageRow in ImageMatrix:
        ProcessedImageRow = []
        for pixel in ImageRow:
            ProcessedPixel = []
            r = pixel[0]
            s = round(c * (1 / np.log(r + 1.01)))
            for i in range(0, 3):
                ProcessedPixel.append(s)
            ProcessedImageRow.append(ProcessedPixel)
        ProcessedImageMatrix.append(ProcessedImageRow)
    ProcessedImage = np.array(ProcessedImageMatrix)
    
    return ProcessedImage
```

```python
ProcessedImage = InverseLogTransform(img, c)
```

### Transformed Grayscale image

```python
ImageDisplayFunc(ProcessedImage, (16, 12))
```
