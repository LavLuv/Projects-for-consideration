### Aim: To segment a grayscale image into various objects separated by edges 

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

def ImageDisplayFunc(Image):
    plt.figure(figsize = (16, 12))
    plt.imshow(Image, cmap = 'gray')
    plt.show()
    
img = cv.imread('C:\\Users\\ASHOK\\Desktop\\Python files\\Image Processing\\Images\\building image.jpg')

GrayscaleImg = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

ImageDisplayFunc(GrayscaleImg)
```

## Using Canny Edge Detection

```python
ProcessedImage = cv.Canny(GrayscaleImg, 100, 140)

ImageDisplayFunc(ProcessedImage)
```
