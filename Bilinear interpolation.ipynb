### Doubling image resolution

```python
def BilinearInterpolationImageResolutionDoubling(image):
    ImageDimensions = image.shape

    b = ImageDimensions[0]
    l = ImageDimensions[1]

    ImageMatrix = list(image)

    ManipulatedImageMatrix = []

    for i in range(0, b-1):    
        ManipulatedImageRow = []
        ImageRow = ImageMatrix[i]
        for j in range(0, l-1):
            ImagePixel = ImageRow[j]
            ManipulatedImageRow.append(ImagePixel)
            NextImagePixel = ImageRow[j+1]
            ManipulatedImagePixel = []
            for k in range(0, 3):
                RGBvalue = round((ImagePixel[k] + NextImagePixel[k]) / 2)
                ManipulatedImagePixel.append(RGBvalue)
            ManipulatedImageRow.append(ManipulatedImagePixel)
        ManipulatedImageRow.append(ImageRow[l-1])
        ManipulatedImageRow.append(ImageRow[l-1])
        ManipulatedImageMatrix.append(ManipulatedImageRow)
        
        # 2nd row addition
        
        ManipulatedImageRow = []
        NextImageRow = ImageMatrix[i+1]
        for j in range(0, l-1):
            ImagePixel = ImageRow[j]
            NextImagePixel = NextImageRow[j]
            ManipulatedImagePixel = []
            for k in range(0, 3):
                RGBvalue = round((ImagePixel[k] + NextImagePixel[k]) / 2)
                ManipulatedImagePixel.append(RGBvalue)
            ManipulatedImageRow.append(ManipulatedImagePixel)
            RightImagePixel = ImageRow[j+1]
            RightNextImagePixel = NextImageRow[j+1]
            ManipulatedImagePixel = []
            for k in range(0, 3):
                RGBvalue = round((ImagePixel[k] + NextImagePixel[k] + RightImagePixel[k] + RightNextImagePixel[k]) / 4)
                ManipulatedImagePixel.append(RGBvalue)
            ManipulatedImageRow.append(ManipulatedImagePixel)
        
        ImagePixel = ImageRow[l-1]
        NextImagePixel = NextImageRow[l-1]
        ManipulatedImagePixel = []
        for k in range(0, 3):
            RGBvalue = round((ImagePixel[k] + NextImagePixel[k]) / 2)
            ManipulatedImagePixel.append(RGBvalue)
        ManipulatedImageRow.append(ManipulatedImagePixel)
        ManipulatedImageRow.append(ManipulatedImagePixel)
        ManipulatedImageMatrix.append(ManipulatedImageRow)
        
    # last row addition

    ImageRow = ImageMatrix[b-1]
    ManipulatedImageRow = []
    for j in range(0, l-1):
        ImagePixel = ImageRow[j]
        ManipulatedImageRow.append(ImagePixel)
        NextImagePixel = ImageRow[j+1]
        ManipulatedImagePixel = []
        for k in range(0, 3):
            RGBvalue = round((ImagePixel[k] + NextImagePixel[k]) / 2)
            ManipulatedImagePixel.append(RGBvalue)
        ManipulatedImageRow.append(ManipulatedImagePixel)
    ManipulatedImageRow.append(ImageRow[l-1])
    ManipulatedImageRow.append(ImageRow[l-1])
    ManipulatedImageMatrix.append(ManipulatedImageRow)
    ManipulatedImageMatrix.append(ManipulatedImageRow)


    ProcessedImage = np.array(ManipulatedImageMatrix)
    
    return ProcessedImage
```

```python
ProcessedImage = BilinearInterpolationImageResolutionDoubling(img)

print('-> Dimensions of the new image:-')
print('')

print(ProcessedImage.shape)
```

```python
ImageDisplayFunc(ProcessedImage, (16, 10))
```

### Halving image resolution

```python
ImageMatrix = list(img)

ManipulatedImageMatrix = []

for i in range(0, 319):
    ManipulatedImageRow = []
    ImageRow = ImageMatrix[2*i]
    for j in range(0, 496):
        ImagePixel = ImageRow[2*j]
        ManipulatedImageRow.append(ImagePixel)
    ManipulatedImageMatrix.append(ManipulatedImageRow)

ProcessedImage = np.array(ManipulatedImageMatrix)

print('-> Dimensions of the new image:-')
print('')

print(ProcessedImage.shape)
```

```python
ImageDisplayFunc(ProcessedImage, (16, 10))
```
